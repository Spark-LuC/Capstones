```python
import requests
import pandas as pd
import time
from urllib.parse import quote
```

```python
API_KEY = "HDEV-907006be-6b96-44d1-bff6-ade257f217ff"
headers = {"Authorization": API_KEY}
BASE = "https://api.henrikdev.xyz/valorant"
regions = ["eu", "na", "latam", "br", "ap", "kr"]

for region in regions:
    players_list = []
    url = f"{BASE}/v3/leaderboard/{region}/pc?size=6000"
    r = requests.get(url, headers=headers)
    data = r.json()

    players = data["data"]["players"]   
    for p in players:
        players_list.append({
            "name": p["name"],       
            "tag": p["tag"],
            "region": region,
            "rr": p["rr"],
            "rank": p["leaderboard_rank"],
            "puuid": p["puuid"]
        })

    print(f"{region} collected")
    players_df = pd.DataFrame(players_list)
    players_df.to_csv(f"players_{region}.csv", index=False)
    print(f"CSV for {region} saved")

    time.sleep(3)
```
